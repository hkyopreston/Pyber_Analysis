```python
# Import Dependencies with an alias of 'plt'

import matplotlib.pyplot as plt
```

```python
# Add Data

#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a plot using MATLAB approach

plt.plot(x_axis, y_axis)
```

```python
# Create a plot using object-oriented interface method (ax.plt())

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add a grid

plt.grid()

# Add the legend.

plt.legend()
```

```python
# Skill Drill

# # Create the plot and add a label for the legend using object oriented method.

ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.

ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.

ax.set_ylim(0, 45)

# Create a title.

ax.set_title("PyBer Fare by Month")

# Add a grid

ax.grid()

# Add the legend.

ax.legend()
```

# Create Pie Chart using object-oriented interface

```python
# Skill Drill

# # Create the Chart and increase the figure size to 8x8
fig, ax = plt.subplots(figsize=(8,8))

# Add new colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Explode the two highest percentage months
explode_values = (0, 0, .2, 0, 0, 0, .25, 0, 0, 0, 0, 0) 

ax.pie(y_axis, 
       labels=x_axis, 
       explode=explode_values, # activate the explode variable
       colors=colors, # activate the colors variable
       shadow=True, # Add a shadow
       counterclock=False, # Reverse the order so that month is clockwise
       startangle=100, # Add a start angle so that January is at the top
       autopct='%1.1f%%') # Add Percentage to each wedge of pie


plt.show()
```

```python
# Line chart with error bars using MATLAB approach

%matplotlib inline
```

```python
# Import dependencies

import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis. 

stdev = statistics.stdev(y_axis)
stdev
```

```python
# Chart standard deviations using errorbars() method

plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# Add a "cap" to the error bars that's equal to the scalar value

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Adding error bars and a capsize using the object-oriented approach

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add error bars and caps to a bar chart using MATLAB approach

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add error bar and caps to a bar chart using object-oriented approach

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Adjust the major x-axis ticks on a horizontal bar chart using MATLAB

import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Adjust the major x-axis ticks on a horizontal bar chart using object-oriented method
# You can also use ax.xaxis.set_ticks() instead of ax.set_xticks()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# add minor ticks without numerical value between major ticks - object-oriented approach

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
