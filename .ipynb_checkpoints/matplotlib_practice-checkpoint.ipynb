```python
# Import Dependencies with an alias of 'plt'

import matplotlib.pyplot as plt
```

```python
# Add Data

#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a plot using MATLAB approach

plt.plot(x_axis, y_axis)
```

```python
# Create a plot using object-oriented interface method (ax.plt())

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().
```

```python
# Create the plot and add a label for the legend.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Create a title.

plt.title("PyBer Fare by Month")

# Add a grid

plt.grid()

# Add the legend.

plt.legend()
```

```python
# Skill Drill

# # Create the plot and add a label for the legend using object oriented method.

ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.

ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.

ax.set_ylim(0, 45)

# Create a title.

ax.set_title("PyBer Fare by Month")

# Add a grid

ax.grid()

# Add the legend.

ax.legend()
```

# Create a Vertical Bar Chart

```python
# Create the plot

plt.bar(x_axis, y_axis, color ='green', label ='Boston')

# Create labels for the x and y axes. 

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title

plt.title("PyBer Fare by Month")

# Add the legend. 

plt.legend()
```

# Create a Horizontal Bar Chart

```python
# Create the plot

plt.barh(x_axis, y_axis)
```

```python
# If you want data on the opposite axes, switch the arguments in the barh() function: 

plt.barh(y_axis, x_axis)
```

```python
# To invert the y-axis to have the months in ascending order, use the gca() method

# Create the plot, color = magenta, label = Boston, Add Title to x and y axis, Add a Title

plt.barh(x_axis, y_axis, color = 'magenta', label = 'Boston')
plt.gca().invert_yaxis()

plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
# Skill Drill - using the MATLAB plotting approach, make the following changes. 

    # Change the color of the markers to red.
    # Add a legend for the city of Chicago.
    # Add a title and axes labels.
    # Switch the axis so that the Fare($) data is on the x-axis.
    # Add a limit for the x-axis data.
    # Invert the y-axis so that January is at the top. 
    
    
plt.scatter(y_axis, x_axis, color='red', label = 'Chicago')
plt.gca().invert_yaxis()

plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')

plt.xlim(0,50)


plt.legend()
```

```python
# Create a bubble chart

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjust the size by multiplying the data in the y-axis by any value. 
# Iterate through the y-axis data and multiply each data point by 3 and add it to a new list

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Use the new y-axis list for the s parameter:

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# The Format for List comprehension is as follows

# new_list = [expression for item in list if conditional]

# Uee the list comprehensin technique to multply each data point in the y-axis list. 

plt.scatter(x_axis, y_axis, s=[i *3 for i in y_axis])
```

```python
# Using the object-oriented approach, make the following changes to your bubble chart:

    # Change the color of the markers to sky blue.
    # Change the size of the markers to 5 times each data point.
    # Make the color 20% transparent.
    # Add a black edge color to the circles.
    # Make the linewidth of the circles 2 points.
    # Add a legend for the city of Boston.
    # Add a title.
    # Switch the axis so that the Fare($) data is on the x-axis.
    # Add a limit for the x-axis data.
    # Invert the y-axis so that January is at the top
```

```python
# Create a Buble Chart using the object-oriented interface approach

fig, ax = plt.subplots()
plt.gca().invert_yaxis()

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*5)
    
ax.scatter(y_axis, x_axis, s=y_axis_larger, c='xkcd:sky blue', alpha = 0.2, edgecolors = 'black', linewidth = 2, label = 'Boston')

ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")

ax.set_xlim(0,50)

ax.legend()
```
